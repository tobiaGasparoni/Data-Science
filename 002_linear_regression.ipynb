```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('1.01.+Simple+linear+regression.csv')
```

```python
data
```

```python
data.describe()
```

Why would I predict the GPA with SAT?
1. SAT is a good estimator for intellectual capacity and capability
2. Almost all colleges care about the SAT
3. The SAT stood the test of time

```python
y = data['GPA']
x1 = data['SAT']
```

```python
plt.scatter(x1, y)
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.show()
```

```python
x = sm.add_constant(x1)
# OLS: Ordinary Least Squares regression
results = sm.OLS(y, x).fit()
results.summary()
```

Base equation: y = b_0 + b_1 * x_1

3 tables:

### Model summary

Ordinary Least Squares (OLS): the most common method to estimate the linear regression equation.

Least squares: minimum squares error or SSE.

Lower error => better explanatory power

Other methods:

- Generalized least squares
- Maximum likelihood estimation
- Bayesian regression
- Kernel regression
- Gaussian process regression

### Coefficients tables

const: b_0 is 0.275

SAT: b_1 is 0.0017

Regression equation: y = 0.275 + 0.0017 * SAT

The lower the std error, the better the estimate.

t statistic

p > |t|

The null hypothesis in this test is H_0: B=0 

Interpretation of null hypothesis:

1. Is the coefficient equal to 0?
2. Is this a useful variable? If the coefficient is 0, it is useless since change in x_1 doesn't affect anything.
3. Does it help explain the variability we have in this case?
4. A p-value < 0.05 means that the variable is significant
5. A p-value of 0.000 is ideal
6. Conclusion: SAT is a significant variable when predicting college GPA.
